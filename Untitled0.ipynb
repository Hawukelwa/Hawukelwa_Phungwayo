```python
import pandas as pd

# List files in the content folder
import os
os.listdir("/content")
```

```python
df = pd.read_excel(
    "/content/20251208 HigherHealth_Stage2 (1500) 1017_20251210.xlsx",
    sheet_name="3 - Learner Data & Achievement"
)

df.head()
```

```python
df.columns
```

```python
df["National ID"] = df["National ID"].astype(str)

df["National ID"].head()
```

```python
import string
import random

def generate_certificate(id_number, sequence, year_prefix="25"):
    id_number = str(id_number)

    if len(id_number) < 6:
        return None

    # First 6 digits of ID
    first_6_digits = id_number[:6]

    # Deterministic random letter based on ID
    random.seed(id_number)
    random_letter = random.choice(string.ascii_uppercase)

    # Sequence starting from 06001
    sequence_number = 6000 + sequence
    sequence_str = str(sequence_number).zfill(5)

    return f"{year_prefix}{first_6_digits}{random_letter}{sequence_str}"
```

```python
df = df.reset_index(drop=True)

df["SEQUENCE_INDEX"] = df.index + 1

df["QCTO_CERT_NUMBER"] = df.apply(
    lambda row: generate_certificate(
        row["National ID"],
        row["SEQUENCE_INDEX"]
    ),
    axis=1
)
```
